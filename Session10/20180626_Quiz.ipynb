# 1.oracle kopo 서버2에서 kopo_product_volume 불러오세요.

```python
# PANDAS 패키지 불러오기
import pandas as pd 
import pymysql 
from sqlalchemy import create_engine 
 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
storeData = pd.read_sql_query('SELECT * FROM KOPO_PRODUCT_VOLUME', engine) 
storeData
```

# 2.컬럼을 모두 대문자로 변경하세요.

```python
storeData.columns = ['REGIONID','PRODUCTGROUP','YEARWEEK','VOLUME']
print(storeData)
```

# 3.kopo_product_volume 에서 지역 ST0002 상품이면서 거래량이 30만건 이상인 부분만 추출하세요.

```python
# 조건 설정
answer = storeData.query('PRODUCTGROUP=="ST0002" & VOLUME > 300000')
print(answer)
```
