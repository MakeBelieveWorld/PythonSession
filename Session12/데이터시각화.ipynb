# 2차원 차트 Subplot

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

plt.figure(figsize=(7,5))
plt.subplot(211)
plt.plot(y[:, 0], 'b',lw=1.5, label = '1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

#첫번째 컬럼자료를 첫번째(왼쪽) 축에 대응하여 그림
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label = '1st')
plt.plot(y[:, 0], 'bo')
plt.grid(True)
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
#두번째 컬럼자료를 두번쨰(오른쪽) 축에 대응하여 그림
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw=1.5, label = '2nd')
plt.plot(y[:, 1], 'go')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

# 차트 종류 변경

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 150개의 2열짜리 정규분포 생성
y=np.random.standard_normal((150,2))

plt.figure(figsize=(7,5))
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')
plt.scatter(y[:,0],y[:,1],marker='o', edgecolors='b')
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 정규분포 50개행 컬럼2개
y=np.random.standard_normal((50,2))
# 0~10까지 랜덤숫자 50개
c= np.random.randint(0,10,len(y))

plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],marker='o',c=c)
plt.colorbar()

plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')
```

```python
#객체 차트
# 빈캔버스에 추가함
fig = plt.figure()
# 위치 및 너비/높이

# 1~5까지 11개의 점 삽입
x = np.linspace(1,5,11)
y = x**2

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax1.plot(x , y)
ax2.plot(y , x)
ax1.set_title('plot1')
ax2.set_title('plot2')
```

```python
#길이크기조정, 해상도 조정

fig = plt.figure(figsize = (8,5), dpi = 60)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```
